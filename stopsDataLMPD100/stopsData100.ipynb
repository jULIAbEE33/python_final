```python
df1 = pd.read_csv("stopsData100.csv")
print(df1)
```

```python
#create sqlite db
db = sqlite3.connect("stops_data.db")

#convert pandas DF to sql table
df1.to_sql("stops_data", db, if_exists="replace")

#view DF
def run_query(query):
    return pd.read_sql_query(query,db)
```

```python
query = "select * from stops_data;"
run_query(query)
```

```python
# get information about my staging table 
query_table_help = """
SELECT name , *FROM PRAGMA_TABLE_INFO('stops_data');
"""
run_query(query_table_help)
```

```python
#create incident counts per division

query_map = """
SELECT 
ACTIVITY_DIVISION
,count(*) as "Incidents"
from stops_data
group by ACTIVITY_DIVISION
order by "ACTIVITY_DIVISION" asc;

"""
run_query(query_map)
```

```python
query_race = """
SELECT 
DRIVER_RACE
,count(*) as "Incidents"
from stops_data
group by ACTIVITY_DIVISION
order by "ACTIVITY_DIVISION" asc;

"""
run_query(query_race)
```

```python
query_gender = """
SELECT 
DRIVER_GENDER
,count(*) as "Incidents"
from stops_data
group by ACTIVITY_DIVISION
order by "ACTIVITY_DIVISION" asc;

"""
run_query(query_gender)
```

```python
query_driver_age = """
SELECT 
DRIVER_AGE_RANGE
,count(*) as "Incidents"
from stops_data
group by ACTIVITY_DIVISION
order by "ACTIVITY_DIVISION" asc;

"""
run_query(query_driver_age)
```

```python
query_driver_race2 = """
SELECT 
DRIVER_RACE
,count(*) as "Incidents"
from stops_data
group by DRIVER_RACE
order by "DRIVER_RACE" asc;

"""
run_query(query_driver_race2)
```

```python
query_driver_demo = """
SELECT 
DRIVER_GENDER as "Gender", DRIVER_RACE as "Race", DRIVER_AGE_RANGE as "Age Range"
,count(*) as "Incidents"
from stops_data
group by DRIVER_AGE_RANGE
order by "DRIVER_AGE_RANGE";

"""
run_query(query_driver_demo)
```
